```python
def plot_graph(df,graph_type,col_1=None,col_2=None):
    print("############ --- plot garph funtion called --- ############")
    print(df.head(2))
    print(graph_type)
    print(col_1)
    print(col_2)
    
    if df[col_1].dtype == "float64" or "int64" and df[col_2].dtype == "float64" or "int64":
        print("int")
    else:
        print("cat")
        
    if col_1 and col_2 is not None:
        
        print("mulitvarite anylisis")
        if graph_type == "scatterplot":
            print("scatterplot")
            scatterplot = sns.scatterplot(data=df,x=df[col_1],y=df[col_2])
            fig = scatterplot.get_figure()
            fig.savefig("scatterplot.png")
            fig.show()
        else:
            print("line_plot")
            lineplot = sns.lineplot(data=df, x=col_1,y=col_2)
            fig = lineplot.get_figure()
            fig.savefig("lineplot.png")
            lineplot.show()


    else:
        print("bivariate anylisis")
        if graph_type == "histplot":
            print("####### {graph_type } #######")
            sns.histplot(data=df, x=col_1)
            fig = histplot.get_figure()
            fig.savefig("histplot.png")
            histplot.show()
        else:
            sns.countplot(x=df[col_1])
            fig = countplot.get_figure()
            fig.savefig("countplot.png")
            countplot.show()

            
```

```python

import seaborn as sns
import os
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

def scatterplot(df, col_1, col_2):
    if os.path.exists("./static/img/plot_imgs/scatterplot.jpeg"):
        os.remove("./static/img/plot_imgs/scatterplot.jpeg")
        sns.scatterplot(data=df, x=df[col_1], y=df[col_2])
        plt.xlabel(col_1)
        plt.ylabel(col_2)
        plt.savefig("./static/img/plot_imgs/scatterplot.jpeg")
        plt.show()
        pass
    else:
        sns.scatterplot(data=df, x=df[col_1], y=df[col_2])
        plt.xlabel(col_1)
        plt.ylabel(col_2)
        plt.savefig("./static/img/plot_imgs/scatterplot.jpeg")
        plt.show()

    return "./static/img/plot_imgs/scatterplot.jpeg"


def lineplot(df, col_1, col_2):
    if os.path.exists("./static/img/plot_imgs/lineplot.jpeg"):
        os.remove("./static/img/plot_imgs/lineplot.jpeg")
        sns.lineplot(data=df, x=df[col_1], y=df[col_2])
        plt.xlabel(col_1)
        plt.ylabel(col_2)
        plt.savefig("./static/img/plot_imgs/lineplot.jpeg")
        plt.show()
        pass
    else:
        sns.lineplot(data=df, x=df[col_1], y=df[col_2])
        plt.xlabel(col_1)
        plt.ylabel(col_2)
        plt.savefig("./static/img/plot_imgs/lineplot.jpeg")
        plt.show()

    return "./static/img/plot_imgs/lineplot.jpeg"


def countplot(df, col_1):
    if os.path.exists("./static/img/plot_imgs/countplot.jpg"):
        os.remove("./static/img/plot_imgs/countplot.jpg")
        sns.countplot(x=df[col_1])
        plt.savefig("./static/img/plot_imgs/countplot.jpg")
        plt.show()
    else:
        sns.countplot(x=df[col_1])
        plt.savefig("./static/img/plot_imgs/countplot.jpg")
        plt.show()


    return "./static/img/plot_imgs/countplot.jpg"


def histplot(df, col_1):
    if os.path.exists("./static/img/plot_imgs/histplot.jpg"):
        os.remove("./static/img/plot_imgs/histplot.jpg")
        sns.histplot(data=df, x=col_1)
        plt.savefig("./static/img/plot_imgs/histplot.jpg")
        plt.show()
    else:
        sns.histplot(data=df, x=col_1)
        plt.savefig("./static/img/plot_imgs/histplot.jpg")
        plt.show()
    return "./static/img/plot_imgs/histplot.jpg"
```
